```python
import pandas as pd
import requests
import zipfile
import io
import os
```

```python
fips_df = pd.read_excel(
    'https://www2.census.gov/programs-surveys/popest/geographies/2018/all-geocodes-v2018.xlsx',
    header=4, dtype=str)
```

```python
state_fips_df = fips_df[fips_df['Summary Level'] == '040'
                       ][['State Code (FIPS)', 
                         'Area Name (including legal/statistical area description)']
                        ].set_index('Area Name (including legal/statistical area description)')
```

```python
state_fips_dict = state_fips_df.to_dict()['State Code (FIPS)']
```

```python
tl_baseurl_2018 = 'https://www2.census.gov/geo/tiger/TIGER2018/'
```

```python
url = os.path.join(tl_baseurl_2018, 'COUSUB/tl_2018_08_cousub.zip')
local_path = '/tmp/'
print(f'Downloading shapefile {url}...')
r = requests.get(url)
z = zipfile.ZipFile(io.BytesIO(r.content))
# z = zipfile.ZipFile(r.content)
print("Done")
z.extractall(path=local_path) # extract to folder
filenames = [y for y in sorted(z.namelist()) for ending in ['dbf', 'prj', 'shp', 'shx'] if y.endswith(ending)] 
print(filenames)
```

```python
county_mask = fips_df['Summary Level'] == '050'
county_cols = ['County Code (FIPS)',
               'State Code (FIPS)',
               'Area Name (including legal/statistical area description)']

county_fips_df = fips_df[county_mask][county_cols].set_index([county_cols[-2], county_cols[-1]])
```
